##### independent two sample t-test(독립표본 t-검정)
- 두 개의 독립적인 정규분포에서 나온 각각의 데이터셋을 사용하여
- 두 정규분포의 기댓값이 동일한지 검사한다.
- 두 정규분포의 분산값이 같은 경우와 다른경우에 각각 사용하는 검정 통계량이 다름(이건 등분산 검정에서 확인)

** example **
- 두 정규 분포의 기댓값이  $μ_1=0 ,  μ_2=0.5$ 로 다르고 분산은  $ σ_1=σ_2=1$  으로 같으며 샘플의 수가  $ N_1=N_2=10 $ 인 경우를 실행해 보자

```python
N_1 = 10; mu_1 = 0; sigma_1 = 1
N_2 = 10; mu_2 = 0.5; sigma_2 = 1
np.random.seed(0)
x1 = sp.stats.norm(mu_1, sigma_1).rvs(N_1)
x2 = sp.stats.norm(mu_2, sigma_2).rvs(N_2)
sns.distplot(x1, kde=False, fit=sp.stats.norm)
sns.distplot(x2, kde=False, fit=sp.stats.norm)
plt.show()
```

```python
sp.stats.ttest_ind(x1, x2, equal_var=True)
# 귀무가설을 둘의 기댓값이 같다고 설정하고
# 실제로는 다르게하여 검정하였더니 귀무가설 accept(같다고 볼 수 있다.)
# Type 2 error
```

```python
N_1 = 50; mu_1 = 0; sigma_1 = 1
N_2 = 100; mu_2 = 0.5; sigma_2 = 1
np.random.seed(0)
x1 = sp.stats.norm(mu_1, sigma_1).rvs(N_1)
x2 = sp.stats.norm(mu_2, sigma_2).rvs(N_2)
sns.distplot(x1, kde=False, fit=sp.stats.norm)
sns.distplot(x2, kde=False, fit=sp.stats.norm)
plt.show()
```

```python
sp.stats.ttest_ind(x1, x2, equal_var=True)
# 데이터 갯수를 증가시켰더니 p-value가 0.008까지 감소하였음(reject)
```
