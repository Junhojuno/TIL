## F 분포
- 카이제곱 분포를 따르는 두 확률 변수를 각각의 자유도(자료의 수, sampling 수)로 나눠
- 각각의 비율을 비교

```python
%matplotlib inline
```

```python
xx = np.linspace(0.03, 3, 1000)
plt.plot(xx, sp.stats.f(1, 1).pdf(xx), label="F(1,1)")
plt.plot(xx, sp.stats.f(5, 1).pdf(xx), label="F(5,1)")
plt.plot(xx, sp.stats.f(10, 10).pdf(xx), label="F(10,10)")
plt.legend()
plt.show()

# (F(n1,n2))
# n1, n2가 같을 때 1이 아닌 어떤수에서 밀도가 높음(더 흔하게 발생한다는 의미)
```

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 10000

n1 = 10
x1 = rv.rvs((n1, M)) # 원소가 10000개인 행벡터가 10개
t1 = (x1 ** 2).sum(axis=0)

N2 = 10
x2 = rv.rvs((N2, M))
t2 = (x2 ** 2).sum(axis=0)

t = t2 / t1
sns.distplot(t, bins=200, kde=False)
plt.xlim(-0.1, 3)
plt.title("N1=10, N2=10")

plt.show()
```
