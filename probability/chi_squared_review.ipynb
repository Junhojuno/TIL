## 카이제곱 분포
- 정규분포를 따르는 확률변수들의 제곱의 합의 분포를 표현
- 이 또한 모수로 자유도(n)를 갖는데, n이 커질수록 정규분포로 다가간다...

```python
x = np.linspace(0.01, 10, 100)
for df in [1, 2, 3, 4, 5, 6, 9]:
    rv = sp.stats.chi2(df=df)
    plt.plot(x, rv.pdf(x),label=("chi2 (자유도 = %d)" % df))
    
plt.xlim(0, 10.1)
plt.ylim(0, 0.6)
plt.legend()
plt.show()
```

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 1000

plt.subplot(1, 2, 1)
N = 1
x = rv.rvs((N, M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1, 10)
plt.ylim(0, 500)
plt.title("N = 1")

plt.subplot(1, 2, 2)
N = 3
x = rv.rvs((N, M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1, 10)
plt.ylim(0, 500)
plt.title("N = 3")

plt.show()
# N = 1 또는 N = 2일때, 0에서 밀도가 가장 높지만
# N = 3 이 넘어가면 1근처, 2근처...점차 0에서 멀어지고 0근처에서 밀도가 0에 가까워짐.
```
