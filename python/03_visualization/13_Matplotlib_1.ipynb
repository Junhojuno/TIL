```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

#### basic line plot

```python
y = np.random.randint(1,10, size=10)
plt.plot(y)
plt.show()
```

```python
# x 데이터를 추가
x = range(0,100,10)
print(list(x), y)
```

```python
plt.plot(x,y)
plt.show()
```

#### 스타일 설정
- option
    - color
    - marker
    - line
- custom style 

```python
# option
plt.plot(y, 'r*:')
plt.show()
```

```python
# custom
plt.plot(y, c='r', lw=1, ls="-.", marker="*", ms=20, mec="b", mew=5, mfc="y")
plt.show()
```

#### x축, y축 Limit

```python
plt.plot(y)
plt.xlim(-5, 15)
plt.ylim(-10, 20)
plt.show()
```

```python
# 위 그래프의 tick : x tick : -5.0, -2.5....
```

```python
plt.plot(y)
plt.xticks(range(10))
plt.yticks([1,2,3,5])
plt.show()
```

```python
plt.plot(y)
plt.grid(True)
plt.show()
```

```python
# 다중 데이터 표현

y1 = np.random.randint(1,10, size=5)
y2 = np.random.randint(1,10, size=5)
y3 = np.random.randint(1,10, size=5)

x = range(5)
print(list(x), y1, y2, y3)

plt.plot(x, y1, 'ro-', x,y2,'g*-.', x,y3,'b^:')
plt.show()
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

##### 범례 - legend
- label, location

```python
plt.plot(y1, label="y1")
plt.plot(y2, label="y2")
plt.legend(loc=0) # 최적화된 위치에 알아서...
plt.show()
```

##### title 추가

```python
plt.plot(y1)
plt.title("Title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.show()
```

```python
plt.plot(y1)
plt.title("타이틀")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.show()
```

##### Figure : 그래프가 그려지는 영역
- 그래프가 그려지는 영역 하나의 figure에 여러개의 그래프를 그릴 수 있습니다.
- 그래프 크기 조절
- 한꺼번에 여러개의 그래프를 그려보겠습니다.

```python
# 그래프의 크기 조절
plt.figure(figsize=(8, 3)) # (가로, 세로)
plt.plot(y)
plt.show()
```

```python
# 여러개의 그래프 출력
y = y[:5]
print(y, y1, y2, y3)
plt.figure(figsize=(10,8))

plt.subplot(2,2,1) # 2 X 2 행렬에 1번째에 그래프를 그린다.
plt.plot(y)

plt.subplot(2,2,2) # 2 X 2 행렬에 2번째에 그래프를 그린다.
plt.plot(y1)

plt.subplot(2,2,3) # 2 X 2 행렬에 3번째에 그래프를 그린다.
plt.plot(y2)

plt.subplot(2,2,4) # 2 X 2 행렬에 4번째에 그래프를 그린다.
plt.plot(y3)

plt.show()
```
