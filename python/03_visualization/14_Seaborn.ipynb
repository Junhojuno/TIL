````python
def sinplot():
    x = np.linspace(0,10, 101) # 0~10까지 101등분
    for i in range(6):
        plt.plot(x, np.sin(i + x))
        
        
sinplot()
````

##### color palettes

````python
# palette 적용
sns.set_palette("hls",8)
sinplot()
````

````python
colors = ["muted","cool","RdBu","CMRmap"]
for color in colors:
    p = sns.color_palette(color, 8)
    sns.palplot(p)
````

#### sample data
- https://github.com/mwaskom/seaborn-data

````python
iris = sns.load_dataset("iris") # pandas형태로 가져온다
iris.tail()
````

````python
iris.petal_length.values
````

````python
# histogram
sns.rugplot(iris.petal_length.values)
````

````python
sns.kdeplot(iris.petal_length.values)
````

````python
sns.distplot(iris.petal_length.values) # 위의 두 그래프가 같이 나옴
````

````python
sns.distplot(iris.petal_length.values, rug=True) 
````

##### joint plot

````python
sns.jointplot(x="sepal_length",y="sepal_width", data=iris)
````

````python
iris.head()
````

##### pair plot

````python
sns.pairplot(iris, hue="species")
plt.show()
````

##### heatmap

````python
flight = sns.load_dataset("flights")
flight.tail()
````

````python
flight_data = flight.pivot("month","year","passengers")
````

````python
sns.heatmap(flight_data)
````

````python
sns.heatmap(flight_data, cmap="YlGnBu", annot=True, fmt=".2f")
````
