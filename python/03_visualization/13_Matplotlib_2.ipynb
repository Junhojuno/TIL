```python
y = np.random.randint(1, 10, size=5)
x = range(5)

print(list(x), y)
```

```python
plt.bar(x,y)
plt.show()
```

```python
y = np.random.randint(1, 10, size=5)
x = range(5)
error = np.random.randint(1,5, size=5)
print(list(x), y, error)

plt.bar(x,y,yerr=error, alpha=0.5) #alpha : 투명도 / 
# error: 데이터 plot의 끝선에 맞춰 위아래 '+/- 해당 값' 인듯...이거 어디에다 쓰지....  
plt.show()
```

```python
y = np.random.randint(1, 10, size=5)
x = range(5)
error = np.random.randint(1,5, size=5)
print(list(x), y, error)

plt.barh(x,y,xerr=error, alpha=0.5)
plt.show()
```

```python
y1 = np.random.randint(5, 10, size=5)
y2 = np.random.randint(1, 5, size=5)

x = range(5)
# error = np.random.randint(1,5, size=5)
print(list(x), y1, y2)

plt.bar(x,y1)
plt.bar(x,y2)
plt.show()
```

#### pie plot

```python
labels = ["A","B","C","D","E"]
colors = ["y","gold","lightskyblue","lightcoral","r"]


plt.pie(y, labels=labels, autopct="%1.1f%%", colors=colors) # %~% 로 감싸줘야함
plt.show()
```

#### scatter plot
- numpy random rand(uniform), randn(정규분포), randint(균등분포)

```python
y = np.random.rand(500) # 500개 데이터 추출
x = np.random.rand(500)
plt.scatter(x,y)
plt.show() # 점이 균등하게 찍힘
```

```python
v = np.random.randn(500)
w = np.random.randn(500)
plt.scatter(v,w)
plt.show() # 0에 몰려있다.
```

```python
a = np.random.randint(1,100, size=500)
b = np.random.randint(1,100, size=500)
plt.scatter(a,b)
plt.show()
```

```python
# 4차원 그래프 그리기
# print(y1, y2)
y1 = np.random.randint(1, 10, size=20)
y2 = np.random.randint(1, 10, size=20)
colors = np.random.rand(20)
area = np.pi * np.arange(10,50,2) ** 2 # 원의 넓이


plt.scatter(y1, y2, s=area, c=colors, alpha=0.5, marker='o')
plt.show()
```

##### histogram

```python
x = np.random.rand(500)
plt.hist(x, bins=50)
plt.show()
```

```python
v = np.random.randn(500)
plt.hist(v, bins=100)
plt.show()
```
