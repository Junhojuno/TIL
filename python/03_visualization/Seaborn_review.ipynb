```python
def sinplot():
    x = np.linspace(0,10, 101)
    for i in range(6):
        plt.plot(x, np.sin(i + x))
        
sinplot()
```

```python
tips = sns.load_dataset("tips")
tips.tail()
```

```python
tips_sum = tips.pivot_table("tip","time","sex", aggfunc=np.sum)
tips_sum
```

```python
# annot : 수치 정보 출력, fmt : 수치데이터 포멧 ".2f" 소수 둘째 자리 ,
# as_cmap: 스무스하게 표현
sns.heatmap(tips_sum, cmap=sns.light_palette("hotpink", as_cmap=True), annot=True, \
           fmt=".2f")
```

```python
tips_mean = tips.pivot_table("tip","time","sex", aggfunc=np.mean)
tips_mean
```

```python
sns.heatmap(tips_mean, cmap=sns.light_palette("seagreen", as_cmap=True), \
           annot=True, fmt='.2g')
```
