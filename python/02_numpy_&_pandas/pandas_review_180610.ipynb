##### apply

```python
def count_char(name):
    return "{}({})".format(name, len(name))

# 새로운 컬럼명, apply 적용 컬럼명, apply할 함수명
df3["Name_count"] = df3["Name"].apply(count_char)
df3
```

##### append

```python
import random, string
```

```python
def get_name():
    names = ["Kim","Lee","Park","Choi","Kang","Hwang","SIn","Bae","Jeon","Jin"]
    return random.choice(names)

def get_age(start=20, end=60):
    return random.randint(start, end+1)

def make_data(rows=10):
    datas = []
    for _ in range(rows):
        data = {"Name":get_name(),"Age":get_age()}
        datas.append(data)
    return datas   
```

```python
d4 = make_data()
df4 = pd.DataFrame(d4) 
df4
```

```python
d5 = make_data()
df5 = pd.DataFrame(d5)
df5
```

```python
# append
df6 = df4.append(df5)
df6
```

```python
# reset_index : index새로 만들기
# drop : 새로 생성되는 index 컬럼 삭제(전에 난잡한 index 제거)
# inplace : 함수를 사용하는 객체 자체 인덱스를 리셋
```

```python
df6.reset_index(drop=True, inplace=True)
df6
```

#### groupby

```python
gdf = pd.DataFrame(make_data(20))
gdf.tail(3)
```

```python
# unique
result1 = gdf['Name'].unique()
result1, len(result1)
```

```python
# groupby의 size
# 'Name'컬럼을 기준으로 묶고, 같은 이름인 사람이 몇명인지를 size로 나타냄
gdf.groupby("Name").size()
```

```python
# 'Name'으로 groupby & 이름을 세고 column명까지 지정
result_df = gdf.groupby("Name").size().reset_index(name='name_counts')
result_df
```

```python
# sorting
# ascending = False : 내림차순
result_df = result_df.sort_values(by=["name_counts"], ascending=False)
result_df
```

```python
# reset_index()필요
result_df.reset_index(drop=True, inplace=True)
result_df
```

### merge

```python
user_df = pd.DataFrame(columns=["UserID","Name","Age"])
for idx in range(1,9):
    name = get_name()
    
    # 중복 이름 제거
    while name in list(user_df["Name"]):
        name = get_name()
        
    # 데이터 name_df insert
    data = {"Name":name, "UserID":idx, "Age":get_age()}
    user_df.loc[len(user_df)] = data
    
user_df
```

```python
money_df = pd.DataFrame(columns=["ID","Money"])

for idx in range(15):
    money = random.randint(1,20) * 1000
    data = {"Money":money, "ID":random.randint(1,8)} # randint는 마지막 숫자를 포함함...ㅡㅡ
    money_df.loc[len(money_df)] = data

money_df
```

```python
# money 데이터 기준으로 merge
money_df.merge(user_df, left_on="ID", right_on="UserID")
```

```python
# user 데이터 기준으로 merge
user_df.merge(money_df, left_on="UserID",right_on="ID")
```

```python
# 만약 두 dataframe의 컬럼명이 같다면?
user_df.rename(columns={"UserID":"ID"}, inplace=True)
result_df = money_df.merge(user_df)
result_df
```

```python
# Name으로 groupby하고 
money_list = result_df.groupby("Name").sum()["Money"].reset_index()
money_list
```

```python
# merge - outer
# 기본적으로 merge는 inner가 default
```

```python
# inner merge
result = pd.merge(user_df, money_list)
result
```

```python
result2 = pd.merge(user_df, money_list, how='outer')
result2
```

```python
# NaN을 0으로 바꾸고 싶다면?
result3 = pd.merge(user_df, money_list, how='outer').fillna(value=0)
result3
```
