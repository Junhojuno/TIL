## Database loading file을 저장하는 방식

### 1. XLS persistence
- db에서 불러온 데이터를 다음에 또 불러오는게 번거롭기 때문에 따로 엑셀 file로 저장
- DataFrame의 엑셀 추출 코드
- xls엔진으로 openpyxls 또는 XlsxWrite 사용

```python
writer  = pd.ExcelWriter('../code/ch5/data/df_routes.xlsx', engine='xlsxwriter')
df_routes.to_excel(writer, sheet_name='Sheet1')
```

### 2.Pickle persistence
- 가장 일반적인 python 파일 persistence
- to_pickle, read_pickle 함수 사용

```python
df_routes.to_pickle("../code/ch5/data/df_routes.pickle")
```

```python
df_routes_pickle = pd.read_pickle("../code/ch5/data/df_routes.pickle")
df_routes_pickle.tail()
```
