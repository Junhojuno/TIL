## Basic functions & operation
- pyplot 객체를 사용하여 데이터를 표시
- pyplot 객체에 그래프들을 쌓은 다음 show로 flush
    - ```import matplotlib.pyplot as plt ```

````python
%matplotlib inline
````

````python
X = range(100)
Y = [value**2 for value in X]
plt.plot(X,Y)
plt.show()
````

##### matplotlib
- Graph는 원래 figure 객체에 생성됨.
- pyplot객체 사용시, 기본 figure에 그래프가 그려짐

````python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

# 하나의 figure에 들어간다.
plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
````

##### figure를 나눠보자.

````python
fig = plt.figure(figsize=(10,5)) # figure 크기 지정

# 두 개의 plot 생성
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(X_1,Y_1, c="b")
ax2.plot(X_2,Y_2, c="r")

plt.show()
````

##### set color

````python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value * 2 for value in X_2]
````

````python
plt.plot(X_1,Y_1, color="#000000")
plt.plot(X_2,Y_2, color="b")
plt.show()
````

##### set linestyle

````python
plt.plot(X_1,Y_1, color="#000000", linestyle="dashed")
plt.plot(X_2,Y_2, color="b", ls="dotted")
plt.show()
````

##### set title

````python
plt.plot(X_1,Y_1, color="#000000", linestyle="dashed")
plt.plot(X_2,Y_2, color="b", ls="dotted")

plt.title("Two lines")
plt.show()
````

````python
# Latex 타입으로 표현 가능
plt.plot(X_1,Y_1, color="#000000", linestyle="dashed")
plt.plot(X_2,Y_2, color="b", ls="dotted")

plt.title(' $y = \\frac{ax + b}{test}$ ')
plt.show()
````

##### set legend(범례)

````python
plt.plot(X_1,Y_1, color="r", linestyle="dashed", label='line 1')
plt.plot(X_2,Y_2, color="b", ls="dotted", label='line 2')

plt.legend(shadow=True, fancybox=True, loc="lower right")
plt.show()
````

##### set grid & x,ylim

````python
plt.grid(True, lw=1.4, ls='--', c='.90')
plt.xlim(-100, 200)
plt.ylim(-200, 200)
````

##### save figure

````python
plt.plot(X_1,Y_1, color="r", linestyle="dashed", label='line 1')
plt.plot(X_2,Y_2, color="b", ls="dotted", label='line 2')

plt.grid(True, lw=1.4, ls='--', c='.90')
plt.xlim(-100, 200)
plt.ylim(-200, 200)

plt.legend(shadow=True, fancybox=True, loc="lower right")
plt.savefig("test.png", c='a')
plt.show()
````
