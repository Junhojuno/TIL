```python
# 데이터를 가져와서 pandas의 dataframe으로 만드는 함수
def get_data(url):
    response = requests.get(url) # get방식 데이터 요청
    # json object로 parsing
    json_info = response.json()
    # json의 dict타입 데이터에서 데이터 가져오기
    companys = json_info['result']['itemList']
    
    df = pd.DataFrame(columns=["종목","시세","전일비","등락율","시가총액","거래량"])
    
    for company in companys:
        df.loc[len(df)] = {
            "종목": company["nm"],
            "시세": company["nv"],
            "전일비": company["cv"],
            "등락율": company["cr"],
            "시가총액": company["mks"],
            "거래량": company["aq"],
        }
    return df
```

### dark sky api
- 위도와 경도를 입력하면 날씨정보를 보내주는 api

```python
FORECAST_TOKEN = "1da567ab436e650c7af58441bd4bdbae"
```

```python
def forecast(lat, lng):
    url = "https://api.darksky.net/forecast/{}/{},{}".format(FORECAST_TOKEN, lat, lng)
    response = requests.get(url)
    json_info = response.json()
    return json_info['timezone'], json_info['hourly']['summary']
```

```python
lat = 37.512
lng = 126.954
forecast(lat, lng)
```

### forecastio를 이용하여 날씨 데이터 가져오기

```python
def forecast2(lat, lng):
    forecast = forecastio.load_forecast(FORECAST_TOKEN, lat, lng) # forecastio.load_forecast의 객체 생성
    byHourly = forecast.hourly()
    return byHourly.summary
```

```python
lat = 37.512
lng = 126.954
forecast2(lat, lng)
```

### bs4
- 네이버 키워드 랭킹 데이터 크롤링
- 다음 키워드 랭킹 데이터 크롤링

```python
def naver():
    df = pd.DataFrame(columns=["rank","keyword"])
    
    response = requests.get("https://www.naver.com/")
    
    dom = BeautifulSoup(response.content, "html.parser") #encoding값이 utf8이 아니면 BeautifulSoup의 encoding argument 설정해줘야 한다. 
    # select : 여러개의 html 엘리먼트를 셀렉팅할때 사용. 결과로 리스트 데이터를 리턴
    # select_one: 하나의 html 엘리먼트를 셀렉팅할때 사용. 결과로 문자열 데이터를 리턴
    keywords = dom.select(".ah_roll_area > .ah_l > .ah_item")
    
    for keyword in keywords:
        df.loc[len(df)] = {
            "rank" : keyword.select_one(".ah_r").text, #해당 엘리먼트 안의 text 데이터를 뽑는다.
            "keyword" : keyword.select_one(".ah_k").text,
        }
    return df
    
naver_df = naver()
naver_df["keyword"]
```

#### file download : requests
- file의 url을 받아서 해당 url의 file을 다운로드하는 방법입니다.

```python
# youtube 영상 다운로드 - bitly로 url길이 단출
url = "https://bit.ly/2JKy2nD"
title = "apple.mp4"
```

```python
def download(title, download_link):
    response = requests.get(download_link, stream=True) # stream = True를 해야 영상(file 등)을 다운받을 수 있음.
    file_size = 0
    with open(title, "wb") as f:
        for chunk in response.iter_content(chunk_size=1024): # byte
            if chunk:
                file_size += 1024
                f.write(chunk)
    return file_size
```

```python
# youtube 영상 다운로드 - bitly로 url길이 단출
url = "https://bit.ly/2JKy2nD"
title = "apple.mp4"
file_size = download(title, url)
```

```python
str(round(file_size/1024/1024,2)) + "Mb"
```

#### 데이터의 크기
- bit : 0 or 1
- byte : 8 bit
- 1kbyte : 1024byte
- 1Mbyte : 1024kbyte
