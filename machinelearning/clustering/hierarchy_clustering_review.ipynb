```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
np.set_printoptions(precision=5, suppress=True)

np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]], size=[100,])
b = np.random.multivariate_normal([0,20],[[3,1],[1,4]], size=[50,])
x = np.concatenate((a,b),)
print(x.shape)

plt.scatter(x[:,0], x[:,1], s=50)
plt.show()
```

```python
z = linkage(x, method='ward')
z[:20]
```

- 계층적 클러스터링, ward
- [데이터 번호1, 데이터 번호 2, 둘 사이의 거리, 클러스터 안의 원소의 갯수]
- 150, 151...은 새롭게 만들어진 클러스터 --> 그래서 해당 부분 원소의 갯수가 3

```python
idxs = [33,68,62,82,63,98]
plt.figure(figsize=(10,10))
plt.scatter(x[:,0], x[:,1], s=50)
plt.scatter(x[idxs,0], x[idxs,1], s=100, c='r')
plt.show()
```

```python
mpl.rcParams["font.family"] = 'DejaVu Sans'

plt.figure(figsize=(10,30))

ax = plt.subplot(111)
dendrogram(z, leaf_font_size=10, orientation='right')

plt.title("Hierarchy Clustering Dendrogram")
plt.xlabel("distance")
plt.ylabel("sample index")

ax.set_xlim(xmin=0.03)
ax.set_xscale("log")
plt.show()
```
