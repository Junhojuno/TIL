### 2. 당뇨병 데이터
- target 데이터 : 1년 후의 당뇨병 진행도
- feature 데이터 : 나이, 성병, BMI지수, 혈압 및 6개의 혈청 검사 수치

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.concat([pd.DataFrame(diabetes.data, columns=diabetes.feature_names),
               pd.DataFrame(diabetes.target, columns=["progress"])],axis=1)
df.tail()
```

```python
df.describe()
```

```python
sns.pairplot(df[['age','sex','bmi','progress']])
plt.show()
```

sex column은 카테고리 분포를 보임

### 3. 체력 검사 데이터
- 턱걸이, 앉았다 일어나기, 점프의 세가지 운동 능력
- 체중, 허리둘레, 맥박의 세가지 신체 상황

```python
from sklearn.datasets import load_linnerud
linnerud = load_linnerud()
df = pd.concat([pd.DataFrame(linnerud.data, columns=linnerud.feature_names),
               pd.DataFrame(linnerud.target, columns=linnerud.target_names)],axis=1)
df.tail()
```

```python
df.iloc[:, [1,2]].head()
```

iloc[행,열] : 위 예를 보면 '행은 다가져오고, 열은 1(Situps), 2(Jumps) 가져와라

```python
sns.pairplot(df.iloc[:,[0,1,2,-1]])
plt.show()
```
