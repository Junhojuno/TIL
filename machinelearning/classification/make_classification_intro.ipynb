##### iris 데이터

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.DESCR)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
sy = pd.Series(iris.target, dtype='category')
```

```python
sy.cat.rename_categories(iris.target_names)
df['species'] = sy
df.tail()
```

```python
sns.pairplot(df, hue='species')
plt.show()
```
