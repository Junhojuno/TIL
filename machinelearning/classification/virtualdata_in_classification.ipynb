```python
from sklearn.datasets import make_classification
```

```python
X1, y1 = make_classification(n_features=1, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=4)
```

- n_informative : 종속변수와 상관관계가 있는 성분의 수
- n_redundant : 다른 독립변수들의 선형조합으로 나타나는 독립변수(성분)의 수

```python
plt.scatter(X1,y1, marker='o', c=y1, s=100, edgecolors='k', linewidths=2)
plt.show()
```

```python
X2, y2 = make_classification(n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=4)
```
